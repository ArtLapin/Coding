```python
import requests
import time
import logging
from bs4 import BeautifulSoup
import csv

# 로깅 설정
logging.basicConfig(filename="crawl_log.txt", level=logging.INFO)

# 데이터 저장 파일 설정
csv_file = "joseon_records.csv"
with open(csv_file, "w", encoding="utf-8", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["URL", "Content"])  # 헤더 추가

# URL 패턴 정의
wang = [chr(i) for i in range(ord('a'), ord('z') + 1)]
mainsub = ['a', 'b']
jaewi = [f"{x:02d}" for x in range(1, 52)]
month = [f"{x:02d}" for x in range(0, 13)]
yoondal = [0, 1]
day = [f"{x:02d}" for x in range(0, 34)]
page = [f"{x:03d}" for x in range(1, 50)]

base_url = "https://sillok.history.go.kr/id/k"

# 데이터 수집
for w in wang:
    for m in mainsub:
        for j in jaewi:
            for m2 in month:
                for y in yoondal:
                    for d in day:
                        for p in page:
                            # URL 생성
                            url = f"{base_url}{w}{m}_1{j}{m2}{y}{d}_{p}"
                            try:
                                # 요청 전송
                                response = requests.get(url, verify=False)
                                if response.status_code == 200:
                                    # HTML 파싱
                                    soup = BeautifulSoup(response.content, "html.parser")
                                    
                                    # p 태그 안의 텍스트 추출
                                    paragraphs = soup.find_all("p", class_="paragraph")
                                    content = "\n".join([p.get_text(strip=True) for p in paragraphs])
                                    
                                    # 데이터 저장
                                    with open(csv_file, "a", encoding="utf-8", newline="") as file:
                                        writer = csv.writer(file)
                                        writer.writerow([url, content])
                                    
                                    # 로그 저장
                                    logging.info(f"SUCCESS: {url}")
                                else:
                                    logging.warning(f"FAILED: {url} with status {response.status_code}")

                            except Exception as e:
                                # 예외 발생 시 로그 기록
                                logging.error(f"ERROR: {url} - {str(e)}")

                            # 요청 간 딜레이
                            time.sleep(0.5)
```

```python
import pandas as pd

# 파일 경로 설정
file_path = "sejong_data.csv"

# 데이터 불러오기
df = pd.read_csv(file_path)

# 결측치 제거
df = df.dropna()  # 결측치 제거

# 중복 제거
df = df.drop_duplicates(subset=["year", "month", "title"], keep="first")

# 연도 변환 (세 자리 → 두 자리)
df["year"] = df["year"].astype(int).apply(lambda x: x % 100)

# 월 변환
def transform_month(month):
    if month % 10 == 1:  # 윤월
        return f"{month // 10}a"
    else:  # 일반 월
        return month // 10

df["month"] = df["month"].astype(int).apply(transform_month)

# 연, 월 기준 정렬
def sort_month_value(month):
    if isinstance(month, str) and month.endswith("a"):
        base_month = int(month[:-1])
        return base_month + 0.5  # 윤달은 정수 + 0.5로 변환
    return int(month)

df["sort_month"] = df["month"].apply(sort_month_value)
df = df.sort_values(by=["year", "sort_month"]).drop(columns=["sort_month"]).reset_index(drop=True)

# 결과 저장
df.to_csv(file_path, index=False, encoding="utf-8-sig")
print(f"\n정리된 데이터를 저장했습니다: {file_path}")

# 결과 확인
print(df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 상소 키워드 설정
appeal_keywords = [
    "상소", "건의", "청원", "상언", "상달", "진정", "상신", "소장", "청첩", "상고", "청사"
]

# 상소 데이터 필터링
data["is_appeal"] = data["title"].astype(str).apply(lambda x: any(keyword in x for keyword in appeal_keywords))
appeal_data = data[data["is_appeal"]]

# 상소 내용 카테고리화 키워드
appeal_categories = {
    "경제": [
        "세금", "조세", "곡물", "환곡", "무역", "재정", "재정개혁", "토지", "곡식", "쌀",
        "납부", "공납", "상납", "지대", "전세", "부역", "수조권", "징세", "균역법", "소작농"
    ],
    "사회": [
        "구휼", "복지", "민생", "치안", "부패", "행정", "구호", "의무", "권리", "구제",
        "질서", "사면", "단속", "관료", "지방관", "민정", "행정개혁", "백성", "도시"
    ],
    "법률": [
        "법률", "형벌", "재판", "판결", "법제", "소송", "사면", "단속", "법관", "의금부",
        "사헌부", "범죄", "처벌", "수사", "법규", "형법", "조사", "소송비", "재판소"
    ],
    "군사": [
        "군사", "병사", "훈련", "국방", "전쟁", "성곽", "방어", "요새", "군대", "전투",
        "무기", "병력", "전진기지", "군량", "수비", "진법", "침략", "무관", "군졸", "성벽"
    ],
    "교육": [
        "교육", "과거제", "유생", "훈민정음", "학문", "경전", "향교", "서원", "독서", "강독",
        "시험", "진사", "문과", "서적", "교재", "문묘", "성리학", "학습", "학자", "강의"
    ]
}

# 상소 카테고리 분류 함수
def classify_appeal_category(text):
    categories = [category for category, words in appeal_categories.items() if any(word in text for word in words)]
    return categories  # 분류되지 않으면 빈 리스트 반환

# 카테고리 분류 적용
appeal_data["상소_카테고리"] = appeal_data["title"].apply(classify_appeal_category)

# "기타" 및 분류되지 않은 데이터 제거
appeal_data = appeal_data[appeal_data["상소_카테고리"].apply(len) > 0]

# 여러 카테고리에 해당하는 데이터를 확장
expanded_appeal_data = appeal_data.explode("상소_카테고리")

# 연도별 상소 빈도 집계
appeal_counts = expanded_appeal_data.groupby(["year", "상소_카테고리"]).size().unstack(fill_value=0)

# 상소 빈도 확인
print(appeal_counts.head())

# 1. 선 그래프 시각화
appeal_counts.plot(kind="line", figsize=(14, 8))
plt.title("세종 대 연도별 상소 이슈 변화", fontsize=16)
plt.xlabel("연도", fontsize=12)
plt.ylabel("상소 빈도", fontsize=12)
plt.legend(title="상소 카테고리", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.grid(visible=True)
plt.tight_layout()
plt.show()

# 2. 히트맵 시각화
plt.figure(figsize=(12, 8))
sns.heatmap(appeal_counts.T, cmap="YlGnBu", annot=False, cbar=True)
plt.title("세종 대 연도별 상소 이슈 변화 (히트맵)", fontsize=16)
plt.xlabel("연도", fontsize=12)
plt.ylabel("상소 카테고리", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from konlpy.tag import Okt

# 데이터 불러오기
file_path = "sejong_data.csv"
data = pd.read_csv(file_path)

# 특정 연도 데이터 필터링
year = 12  # 예: 12년
filtered_data = data[data["year"] == year]

# 형태소 분석기 초기화
okt = Okt()

# 명사 추출 함수
def extract_nouns(text):
    return [noun for noun in okt.nouns(text) if len(noun) > 1]  # 한 글자 명사 제외

# title 열에서 명사 추출
filtered_data["nouns"] = filtered_data["title"].astype(str).apply(extract_nouns)

# 명사 빈도수 계산
all_nouns = [noun for nouns_list in filtered_data["nouns"] for noun in nouns_list]
noun_counts = Counter(all_nouns)

# 상위 20개 단어 출력
top_nouns = noun_counts.most_common(20)
print(f"{year}년에 가장 많이 언급된 단어: {top_nouns}")


# 1. 워드클라우드 시각화
wordcloud = WordCloud(
    font_path='malgun.ttf',  # Windows: 'malgun.ttf', macOS: '/Library/Fonts/AppleGothic.ttf'
    background_color="white",
    width=800,
    height=400
).generate_from_frequencies(noun_counts)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title(f"{year}년 워드클라우드", fontsize=16)
plt.show()
```

```python
# 국방 관련 키워드 설정
defense_keywords = ["군사", "국방", "훈련", "병력", "병사", "성곽", "방어", "무기", "전쟁", "군대"]

# 국방 관련 데이터 필터링
data["is_defense"] = data["title"].astype(str).apply(lambda x: any(keyword in x for keyword in defense_keywords))
defense_data = data[data["is_defense"]]

# 연도별 국방 논의 빈도 계산
defense_counts = defense_data.groupby("year").size()

# 1. 연도별 국방 논의 빈도 선 그래프
plt.figure(figsize=(14, 6))
plt.plot(defense_counts.index, defense_counts.values, marker="o", linewidth=2)
plt.title("세종 대 연도별 국방 논의 빈도", fontsize=16)
plt.xlabel("연도", fontsize=12)
plt.ylabel("국방 논의 빈도", fontsize=12)
plt.grid(visible=True)
plt.tight_layout()
plt.show()

# 2. 히트맵 시각화 준비 (연도별 데이터를 확장)
defense_years = defense_data.groupby(["year"]).size().reset_index(name="count")
heatmap_data = defense_years.pivot_table(index="year", values="count", aggfunc="sum")

# 히트맵 시각화
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap="YlGnBu", annot=True, fmt="d", cbar=True)
plt.title("세종 대 연도별 국방 논의 빈도 (히트맵)", fontsize=16)
plt.xlabel("연도", fontsize=12)
plt.ylabel("국방 논의 빈도", fontsize=12)
plt.tight_layout()
plt.show()

# 특정 연도에 가장 많은 국방 논의가 이루어진 연도
max_defense_year = defense_counts.idxmax()
max_defense_count = defense_counts.max()
print(f"국방 논의가 가장 많이 이루어진 연도는 {max_defense_year}년이며, 빈도는 {max_defense_count}회입니다.")
```

```python
# 데이터 불러오기
file_path = "sejong_data.csv"
data = pd.read_csv(file_path)

# 윤달 처리 함수
def process_month(month):
    if isinstance(month, str) and month.endswith("a"):  # 윤달일 경우
        return int(month[:-1])  # '8a' -> 8
    try:
        return int(month)  # 일반 월은 그대로 유지
    except ValueError:
        return None  # 변환 실패 시 None 반환

# 월 데이터 정리
data["month"] = data["month"].apply(process_month)
data = data.dropna(subset=["month"])  # 결측값 제거
data["month"] = data["month"].astype(int)  # 정수형 변환

# 오랑캐와 여진, 왜구 키워드 설정
keywords = {
    "오랑캐_여진": ["오랑캐", "여진"],
    "왜구": ["왜구"]
}

# 데이터 필터링 함수
def filter_data(text, keyword_list):
    return any(keyword in text for keyword in keyword_list)

# 데이터 필터링 컬럼 추가
data["is_orangcae_yeojin"] = data["title"].astype(str).apply(lambda x: filter_data(x, keywords["오랑캐_여진"]))
data["is_waegu"] = data["title"].astype(str).apply(lambda x: filter_data(x, keywords["왜구"]))

# 연도별 빈도 계산
orangcae_yeojin_yearly = data[data["is_orangcae_yeojin"]].groupby("year").size()
waegu_yearly = data[data["is_waegu"]].groupby("year").size()

# 월별 빈도 계산
orangcae_yeojin_monthly = data[data["is_orangcae_yeojin"]].groupby("month").size()
waegu_monthly = data[data["is_waegu"]].groupby("month").size()

# 1~12월 범위로 데이터 정렬
months = range(1, 13)
orangcae_yeojin_monthly = orangcae_yeojin_monthly.reindex(months, fill_value=0)
waegu_monthly = waegu_monthly.reindex(months, fill_value=0)

# 1. 연도별 침략 빈도 선 그래프
plt.figure(figsize=(14, 6))
plt.plot(orangcae_yeojin_yearly.index, orangcae_yeojin_yearly.values, label="오랑캐+여진", marker="o", linewidth=2)
plt.plot(waegu_yearly.index, waegu_yearly.values, label="왜구", marker="o", linewidth=2)
plt.title("연도별 오랑캐+여진과 왜구 침략 빈도", fontsize=16)
plt.xlabel("연도", fontsize=12)
plt.ylabel("침략 빈도", fontsize=12)
plt.legend()
plt.grid(visible=True)
plt.tight_layout()
plt.show()

# 2. 월별 선 그래프 시각화
plt.figure(figsize=(12, 6))
plt.plot(months, orangcae_yeojin_monthly, label="오랑캐+여진", marker="o", linewidth=2)
plt.plot(months, waegu_monthly, label="왜구", marker="o", linewidth=2)
plt.title("월별 오랑캐+여진과 왜구 침략 빈도", fontsize=16)
plt.xlabel("월", fontsize=12)
plt.ylabel("침략 빈도", fontsize=12)
plt.xticks(months)
plt.legend()
plt.grid(visible=True)
plt.tight_layout()
plt.show()

```
